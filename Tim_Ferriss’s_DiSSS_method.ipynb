```python
# 1. VARIABLE: We store the user's input in a box called 'name'
name = input("What is your name? ")

# 2. LOGIC: We check what is inside the box
if name == "Tim":
    print("Welcome to the 4-Hour Club!")
else:
    # This happens if the name is NOT Tim
    print("Sorry, " + name + ". Access Denied.")
```
